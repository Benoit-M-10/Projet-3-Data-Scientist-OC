```python
buildingID_toAdd = df_2015[~df_2015['OSEBuildingID'].isin(df_2016['OSEBuildingID'])]['OSEBuildingID']
```

```python
df_final = df_2016.append(df_2015[df_2015['OSEBuildingID'].isin(buildingID_toAdd)], ignore_index=True)
```

```python
index_2015 = df_final[df_final['OSEBuildingID'].isin(buildingID_toAdd)].index
```

```python
df_final.loc[index_2015,'ZipCode'] = df_final.loc[index_2015,'Zip Codes']
df_final.loc[index_2015,'Comments'] = df_final.loc[index_2015,'Comment']
df_final.loc[index_2015,'TotalGHGEmissions'] = df_final.loc[index_2015,'GHGEmissions(MetricTonsCO2e)']
df_final.loc[index_2015,'GHGEmissionsIntensity'] = df_final.loc[index_2015,'GHGEmissionsIntensity(kgCO2e/ft2)']
```

```python
# Pour la colonne 'OtherFuelUse(kBtu)' qui n'est pas présente dans le dataset 2016 on ajoute les données pour les
# bâtiments qui sont dans le dataset 2015.

# On va faire une jointure donc on supprime dans un premier temps la colonne 'OtherFuelUse(kBtu)' de df_final.

df_final = df_final.drop('OtherFuelUse(kBtu)', axis=1)
```

```python
df_final = df_final.merge(df_2015[['OSEBuildingID', 'OtherFuelUse(kBtu)']], how='outer', on='OSEBuildingID')
```

```python
df_final
```

```python
# On remplit maintenant les colonnes 'Latitude' et 'Longitude' des données ajoutées de 2015 en utlisant les informations
# de la colonne 'Location'.

for i in index_2015:
    index_debut_latitude = df_final.loc[i,'Location'].find("'latitude': '")
    index_debut_longitude = df_final.loc[i,'Location'].find("', 'longitude': '")
    index_debut_address = df_final.loc[i,'Location'].find("', 'human_address'")
    
    df_final.loc[i,'Latitude'] = df_final.loc[i,'Location'][index_debut_latitude+13:index_debut_longitude-1]
    df_final.loc[i,'Longitude'] = df_final.loc[i,'Location'][index_debut_longitude+17:index_debut_address-1]
```

```python
df_final['Latitude'] = df_final['Latitude'].astype('float64')
df_final['Longitude'] = df_final['Longitude'].astype('float64')
```
